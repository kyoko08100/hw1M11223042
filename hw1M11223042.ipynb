```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
title = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']
df_Train = pd.read_csv(r'C:\Users\USER\DataMining-hw\hw1\adult.data', names = title, header=None,encoding='unicode_escape')
df_Test = pd.read_csv(r'C:\Users\USER\DataMining-hw\hw1\adult.test', names = title, header=None)
raw_df_train = df_Train.copy()
df_Train = df_Train.drop(columns=['fnlwgt','education-num'])
df_Test = df_Test.drop(columns=['fnlwgt','education-num'])
# 將final weight 去除(因為那是人口普查ID，ID不能被列為Feature)以及education-num
print(df_Train)
print(df_Test)
```

```python
df_Train.replace(" ?", pd.NA, inplace=True)
df_Test.replace(" ?", pd.NA, inplace=True)
print(df_Train)
print(df_Test)
```

```python
label_encoder = LabelEncoder()

# 將類別的相關欄位做Label Encoding
df_Train['workclass'] = label_encoder.fit_transform(df_Train['workclass'].astype('str'))
df_Test['workclass'] = label_encoder.fit_transform(df_Test['workclass'].astype('str'))

df_Train['education'] = label_encoder.fit_transform(df_Train['education'].astype('str'))
df_Test['education'] = label_encoder.fit_transform(df_Test['education'].astype('str'))

df_Train['marital-status'] = label_encoder.fit_transform(df_Train['marital-status'].astype('str'))
df_Test['marital-status'] = label_encoder.fit_transform(df_Test['marital-status'].astype('str'))

df_Train['occupation'] = label_encoder.fit_transform(df_Train['occupation'].astype('str'))
df_Test['occupation'] = label_encoder.fit_transform(df_Test['occupation'].astype('str'))

df_Train['relationship'] = label_encoder.fit_transform(df_Train['relationship'].astype('str'))
df_Test['relationship'] = label_encoder.fit_transform(df_Test['relationship'].astype('str'))

df_Train['race'] = label_encoder.fit_transform(df_Train['race'].astype('str'))
df_Test['race'] = label_encoder.fit_transform(df_Test['race'].astype('str'))

df_Train['sex'] = label_encoder.fit_transform(df_Train['sex'].astype('str'))
df_Test['sex'] = label_encoder.fit_transform(df_Test['sex'].astype('str'))

df_Train['native-country'] = label_encoder.fit_transform(df_Train['native-country'].astype('str'))
df_Test['native-country'] = label_encoder.fit_transform(df_Test['native-country'].astype('str'))

# df_Train['income'] = label_encoder.fit_transform(df_Train['income'].astype('str'))
# df_Test['income'] = label_encoder.fit_transform(df_Test['income'].astype('str'))

print(df_Train)
```

```python
# 將Dataframe分為X(feature)跟Y(class)
train_x = df_Train.iloc[:, 0:12].values
train_y = df_Train.iloc[:, 12].values

test_x = df_Test.iloc[:, 0:12].values
test_y = df_Test.iloc[:, 12].values
print(train_x,train_y,test_x,test_y)
# print(raw_df_train)
```

# CART決策樹 Python

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import confusion_matrix
from six import StringIO

import pydot
import pydotplus
import matplotlib.pyplot as plt
# 訓練CART決策樹
result=[]

# train1 max_depth = 7,max_leaf_nodes=61
model1 = DecisionTreeClassifier(criterion='gini',max_depth=7,max_leaf_nodes=61).fit(train_x,train_y)
result.append(model1.score(train_x,train_y))
# train2 max_depth = 9,max_leaf_nodes=81
model2 = DecisionTreeClassifier(criterion='gini',max_depth=9,max_leaf_nodes=81).fit(train_x,train_y)
result.append(model2.score(train_x,train_y))
# train3 max_depth = 11,max_leaf_nodes=90
model3 = DecisionTreeClassifier(criterion='gini',max_depth=11,max_leaf_nodes=90).fit(train_x,train_y)
result.append(model3.score(train_x,train_y))
# train4 raw
model4 = DecisionTreeClassifier(criterion='gini').fit(train_x,train_y)
result.append(model4.score(train_x,train_y))
model=[model1,model2,model3,model4]
for i in range(4):
    print(result[i])
for i in range(4):
    print(model[i].score(test_x,test_y))

# 使用測試集預測
# model.predict(test_x)
# model.score(test_x,test_y)

# 計算混淆矩陣
cm1 = confusion_matrix(test_y, model1.predict(test_x))
cm2 = confusion_matrix(test_y, model2.predict(test_x))
cm3 = confusion_matrix(test_y, model3.predict(test_x))
cm4 = confusion_matrix(test_y, model4.predict(test_x))
print(cm1)
print(cm2)
print(cm3)
print(cm4)

# 畫出樹的圖到PNG
def print_png(model: str,png_name: str):
    title = ['age','workclass','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country']

    dot_data = StringIO()
    tree.export_graphviz(model,
            out_file=dot_data,
            feature_names=title,
            class_names="10",
            filled=True, rounded=True,
            impurity=False)

    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
    graph.write_png(png_name)
# print_png(model1,r"C:\Users\USER\DataMining-hw\hw1\adult7-61.png")
# print_png(model2,r"C:\Users\USER\DataMining-hw\hw1\adult9-81.png")
# print_png(model3,r"C:\Users\USER\DataMining-hw\hw1\adult11-90.png")
# print_pdf(model4,r"C:\Users\USER\DataMining-hw\hw1\adult-raw.png")
```

# ID3決策樹 R

```python
import os
os.environ['R_HOME'] = r'C:\Users\USER\anaconda3\envs\testhw\Lib\R'


import rpy2.robjects as ro
import matplotlib.pyplot as plt
from rpy2.robjects import pandas2ri
# 要加這個activate，不然無法匯入dataframe到R語言中
pandas2ri.activate()
# 加入 rpart
rpart = ro.r('library(rpart)')

# 直接用 R 語言建立 DataFrame
df_train = ro.DataFrame(df_Train)
df_test = ro.DataFrame(df_Test)
# 在 R 中指定變數
ro.globalenv['train'] = df_train
ro.globalenv['test'] = df_test

# 訓練模型、繪圖、預測(ID3)
rpart_code = """
library(rpart.plot)

my_tree <- rpart(income~., data=train,method="class",control = rpart.control(minsplit=1,cp=0.03))
train_index <- sample(1:32561, size = 32561, replace = FALSE)
income_train <- train$income[train_index]
(train_table <- table(train = train$income[train_index],predict = predict(my_tree, newdata=train, type="class")))
train_acc=sum(diag(train_table))/sum(train_table)

dev.list()
pdf("r-graph.pdf")
dev.cur()
prp(my_tree,faclen=0,fallen.leaves=TRUE,shadow.col="gray",extra=2,cex = 1) 
dev.off()

pred <- predict(my_tree, newdata=test, type="matrix")

"""
ro.r(rpart_code)

# Summary 結果顯示
model_summary = ro.r('summary(my_tree)')
print(model_summary)

# 將 R 語言中的圖表轉換為 Python 語言
pred = ro.r('pred')
train_table = ro.r('train_table')
print(train_table)

# [TP,FN,FP,TN] 計算時將test_y轉為[0,1]
c_matrix=[0,0,0,0]
df_Train['income'] = label_encoder.fit_transform(df_Train['income'].astype('str'))
df_Test['income'] = label_encoder.fit_transform(df_Test['income'].astype('str'))
test_y = df_Test.iloc[:, 12].values
print(df_Train)
for i in range(len(test_y)):
    if int(pred[i][0]) - 1 == test_y[i] and test_y[i] == 1:
        c_matrix[0] += 1
    elif int(pred[i][0]) - 1 != test_y[i] and test_y[i] == 1:
        c_matrix[1] += 1
    elif int(pred[i][0]) - 1 != test_y[i] and test_y[i] == 0:
        c_matrix[2] += 1
    elif int(pred[i][0]) - 1 == test_y[i] and test_y[i] == 0:
        c_matrix[3] += 1
train_acc = ro.r('train_acc')
print("train_acc=",train_acc)
print(c_matrix,end="")
print("test_acc=",(c_matrix[0]+c_matrix[3])/(c_matrix[0]+c_matrix[1]+c_matrix[2]+c_matrix[3]))
```
