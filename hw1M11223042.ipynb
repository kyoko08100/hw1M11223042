```python
import rpy2.robjects as ro

# 加载 rpart 包
rpart = ro.r('library(rpart)')

# 直接用 R 語言建立 DataFrame
df_r = ro.DataFrame({'x': [1, 2, 3], 'y': [4, 5, 6]})

# 在 R 中使用数据
ro.globalenv['my_r_data'] = df_r

# 在 R 中定义 target 变量
rpart_code = """
target <- c(1, 0, 1)
feature1 <- c(1, 2, 3)
feature2 <- c(1, 2, 3)
"""
ro.r(rpart_code)

# 在 R 中执行决策树分析
r_code = """
my_tree <- rpart(target ~ feature1 + feature2, data=my_r_data)
"""
ro.r(r_code)

# 从 R 中提取决策树模型
rpart_model = ro.globalenv['my_r_data']

# 将模型转换为 Python 数据结构
model_summary = ro.r('summary(my_tree)')
print(model_summary)
```
